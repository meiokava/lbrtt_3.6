<font size ="4">Построение графика функции:

$$f(x, y) = 3sin(x)cos(y) - sin(x)cos(y^2)$$

в диапазоне -5 ≤ x ≤ 5 и -5 ≤ y ≤ 5.
    
1. Импорт необходимых библиотек
2. Определение функции f(x, y)
3. Создание диапазонов для x и y
4. Создание сетки из значений x и y
5. Вычисление соответствующего значения z, используя функцию f(x, y)
6. Создание 3D-графика функции, используя функцию plot_surface()

Результирующий график представляет собой трехмерный поверхностный график функции f(x, y) в указанном диапазоне значений x и y.

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return np.sin(x) + np.cos(y)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
